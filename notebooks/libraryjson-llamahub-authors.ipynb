```python
import os
import json
from os import listdir
import toml
```

```python
LLAMA_HUB_PATH = "/Users/nerdai/Projects/llama-hub/llama_hub"
MAPPING_PATH = "/Users/nerdai/Projects/llama_index/llama-index-core/llama_index/core/command_line/mappings.json"
LLAMA_INDEX_PATH = "/Users/nerdai/Projects/llama_index"

with open(MAPPING_PATH, "r") as f:
    mappings = json.load(f)
```

### Loaders

```python
library_path = os.path.join(LLAMA_HUB_PATH, "library.json")
with open(library_path, "r") as f:
    library = json.load(f)
```

```python
SPECIAL_CASES = {
    "KalturaESearchReader": "llama_index.readers.kaltura"
}
```

```python
def index_of_author(lst, author):
    for i, v in enumerate(lst):
        if author in v:
            return i
    return None
```

```python
k = "AirbyteZendeskSupportReader"
obj = library[k]
import_path = mappings[k]
toml_path = os.path.join(
    LLAMA_INDEX_PATH,
    "llama-index-integrations",
    "readers",
    import_path.replace(".", "-").replace("_", "-"),
    "pyproject.toml"
)
pyproject = toml.load(toml_path)

if "author" in obj:
    class_authors = pyproject["tool"]["llamahub"]["class_authors"]
    class_authors[k] = obj["author"]   
```

```python
pyproject
```

```python
for k, obj in library.items():
    if k not in mappings:
        print(f"NOT FOUND: {k}\n")
    if k in mappings:
        
        if k in SPECIAL_CASES:
            import_path = SPECIAL_CASES[k]
        else:
            import_path = mappings[k]
        
        if "llama_index.core" in import_path:
            continue
        
        print(f"{k}")
        # load pyproject.toml
        toml_path = os.path.join(
            LLAMA_INDEX_PATH,
            "llama-index-integrations",
            "readers",
            import_path.replace(".", "-").replace("_", "-"),
            "pyproject.toml"
        )
        pyproject = toml.load(toml_path)

        # add author data
        if "author" in obj:
            class_authors = pyproject["tool"]["llamahub"]["class_authors"]
            class_authors[k] = obj["author"]
            
        # write pyproject.toml
        with open(toml_path, "w") as f:
            toml.dump(pyproject, f)

        # update packages to bump and publish
        with open("packages_to_bump.txt", "a") as f:
            f.write(
                str(os.path.join(
                    LLAMA_INDEX_PATH,
                    "llama-index-integrations",
                    "readers",
                    import_path.replace(".", "-").replace("_", "-")
                )) + "\n"
            )
        print("\n")
```
